# Plot EM, F1

```python
import matplotlib.pyplot as plt


def plot_perf(em, f1, best_epoch, best_em, best_f1, model_name):
    plt.figure(figsize=(9, 6))
    plt.plot(range(len(em)), em, 'o-', alpha=0.5, label='Exact Match (EM-Score)')
    plt.plot(range(len(f1)), f1, 'o-', alpha=0.5,
             label='Partial Match (F1-Score)')
    plt.axvline(x=best_epoch, alpha=0.2,
             label=f'Best Score\n(EM={best_em:.2f}, F1={best_f1:.2f})')
#     plt.plot([best_epoch]*len(em), range(0, 100), '--k', alpha=0.2,
#              label=f'Best Score\n(EM={best_em:.2f}, F1={best_f1:.2f})')
    plt.xticks(size=13)
    plt.yticks(size=13)
    plt.ylim(0, 90)
    plt.xlabel('#Epoch', size=14)
    plt.ylabel('#Performance Score', size=14)
    plt.title(
        f"Performance of the '{model_name}' \ntransformer MRC model on test set",
        size=14
    )
    plt.legend(loc='lower right', fontsize=12)
    plt.show()
```

```python
output_model_dir = '/net/kdinxidk03/opt/NFS/75y/data/qa/output/bart-otquad/'
model_name = 'BASE MODEL + OTQuAD'

with open(f'{output_model_dir}em.txt', 'r') as f:
    em = f.readlines()
with open(f'{output_model_dir}f1.txt', 'r') as f:
    f1 = f.readlines()
    
em = list(map(float, em[0][1:-2].split(',')))
f1 = list(map(float, f1[0][1:-2].split(',')))
best_epoch = em.index(max(em))
best_em = em[best_epoch]
best_f1 = f1[best_epoch]

plot_perf(em, f1, best_epoch, best_em, best_f1, model_name)
print(best_epoch, best_em, best_f1)
```
