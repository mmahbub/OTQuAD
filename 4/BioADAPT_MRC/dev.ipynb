```python
def squad_entries(dataset_name):
  
  dataset = json.load(open(data_path/f'{dataset_name}.json'))
  
  data = dataset['data'][0]
  version = dataset['version']
  para = data['paragraphs']
  title = data['title']
  
  squad_entries = []
  for d in para:
    new_entry = {"qas": [
                 {
                  "id": d['qas'][0]['id'],
                  "question": d['qas'][0]['question'], #"[MASK]"
                  "answers": [""] #d['qas'][0]['answers']
                 }
                 ],
               "context": d['context']
               }

    squad_entries.append(new_entry)
  
  print(len(squad_entries))
  
  squad_json={
            "data": [
              {
                "paragraphs":squad_entries,
                 "title":title
              }],
             "version":version
         }
  
  return squad_json
```
