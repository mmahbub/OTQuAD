```python
fact_data = [x for x in bioasq_test['questions'] if x['type']=='factoid']

para = []

for x in fact_data:
  question = x['body']
  answer = x['exact_answer'][0][0]
  idx = x['id']
  for i,y in enumerate(x['snippets']):
    context = y['text']
    ids = idx+f'_{i:03d}'
    answer_start = context.lower().find(answer.lower())
    if answer_start != -1:
      para.append({'qas': [{'id': ids,
                            'question': question,
                            'answers': [{'answer_start': answer_start, 'text': answer}]}],
                   'context': context})

```
